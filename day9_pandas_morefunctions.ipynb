```python
df.dropna(axis=0, thresh=4)
```

```python
df.fillna(0)
```

```python
df['Age'].fillna(df['Age'].mean())
```

```python
#df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age'].fillna(df['Age'].mean(),inplace = True)
```

```python
df
```

```python
df['Sex']=df[['Sex','Name']].apply(lambda x: 'male' if ('Mr.' in x['Name'])and(pd.isnull(x['Sex'])) else x['Sex'], axis=1)
df
```

```python
df['Sex']=df[['Sex','Name']].apply(lambda x: 'female' if ('Miss' in x['Name'])and(pd.isnull(x['Sex'])) else x['Sex'], axis=1)
df
```

```python
df['Sex']=df[['Sex','Name']].apply(lambda x: 'female' if ('Mrs' in x['Name'])and(pd.isnull(x['Sex'])) else x['Sex'], axis=1)
df
```

Groupby Operations

```python
df = pd.read_csv('Teams.csv')
df
```

```python
df.groupby('Teams')['Points'].mean()
```

```python
df.groupby(['Teams','Year'])['Points'].mean()
```

```python
h = df.groupby(['Teams','Year'])
```

```python
h.groups
```

```python
h.get_group(('Devils',2014))
```

```python
df.groupby('Teams').get_group('Riders')
```

```python
for i,v in df.groupby('Teams'):
    print(i)
    print(v)
```

```python
import numpy as np
```

```python
df.groupby('Teams')['Points'].agg([len, np.mean, np.std, min, max])
```
